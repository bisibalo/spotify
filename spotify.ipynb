```python
songs = songs.rename(columns = {'Unnamed: 0':'id'})
songs = songs.rename(columns={"track_name":"title"})
```

```python
songs = songs[['id','album_name','title','track_genre']]
```

```python
songs
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
songs['tags'] = songs['album_name'] + songs['track_genre']
songs
```

```python
new_df = songs.drop(columns =['album_name','track_genre'])
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 114000, stop_words ='english')
cv
```

```python
vector = cv.fit_transform(new_df['tags'].astype(str))
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector, dense_output=False)
```

```python
similarity
```

```python
new_df[new_df['title']=='To Begin Again'].index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
for i in distance[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
def recommand(songs):
    index = new_df[new_df['title']==songs].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(new_df.iloc[i[0]].title)
```

```python
recommand("To Begin Again")
```
